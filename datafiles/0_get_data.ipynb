# Get data from server

Loads datafiles with batch, microscopy, and FACS data from server

```python
from pathlib import Path
import shutil

server_path = Path('/Volumes/ScientificData/Users/Giulia(botgiu00)/Papers/bottacin2025/Figure_Data/Figure3/')
repo_path = Path.cwd() 

file_list = [
    'HQNO_Calibration/hqno_calibration_microscopy.csv',
    'HQNO_Calibration/hqno_calibration_facs.csv',
    'RHL_Calibration/rhl_calibration_microscopy.csv',
    'RHL_Calibration/rhl_calibration_facs.csv',
    'Batch_Survival/HQNO_RHL_Survival.csv',
    'HQNO_Gradient/hqno_gradient_microscopy.csv',
    'RHL_Gradient/rhl_gradient_microscopy.csv'
    ]

for file_path in file_list:
    src = server_path / file_path
    file_name = Path(file_path).name
    dst = repo_path / file_name  # Explicitly use just the filename

    if src.exists():
        shutil.copy(src, dst)
    else:
        print(f"Server path does not exist: {src}")
```
